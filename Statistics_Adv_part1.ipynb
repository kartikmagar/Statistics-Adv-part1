1. What is a random variable in probability theory?

-> A random variable is a variable that takes on numerical values determined by the outcome of a random experiment. It can be discrete (finite/countable outcomes) or continuous (infinite outcomes in an interval).


2. What are the types of random variables?

-> Discrete Random Variable - takes specific, countable values.
Continuous Random Variable - takes any value within a range or interval.


3. What is the difference between discrete and continuous distributions?

-> Discrete distribution: For discrete variables; probability is assigned to specific values.
Continuous distribution: For continuous variables; probability is given over an interval, and the probability of a single exact value is zero.


4. What are probability distribution functions (PDF)?

-> For discrete variables: PDF gives the probability that a variable equals a particular value (also called PMF).
For continuous variables: PDF is a function whose integral over an interval gives the probability that the variable falls within that interval.

11. What are the characteristics of a normal distribution?

-> Bell-shaped, symmetric about the mean.
Mean = Median = Mode.

12. What is the standard normal distribution, and why is it important?

-> A special normal distribution with mean = 0 and standard deviation = 1.
Used to calculate probabilities using the Z-score and is the basis for many statistical methods.


13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

-> CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution.
It allows the use of normal probability for inference.

-> Point estimate: A single value estimate of a parameter (e.g., sample mean for population mean).
